# Project Water quality
## Subject: Predict if water is safe for human consumption

### Context
Access to safe drinking-water is essential to health, a basic human right and a component of effective policy for health protection. This is important as a health and development issue at a national, regional and local level. In some regions, it has been shown that investments in water supply and sanitation can yield a net economic benefit, since the reductions in adverse health effects and health care costs outweigh the costs of undertaking the interventions.

```python
# load some default Python modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, classification_report, f1_score, accuracy_score, confusion_matrix, precision_recall_fscore_support
from sklearn.model_selection import train_test_split, cross_val_score
plt.style.use('seaborn-whitegrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
# Case rows filling
#0: remove all rows that contains a NaN values, 1: fill the NaN values by interpolation 
CASE=2
# load data
data = pd.read_csv('water_potability.csv')
# display some data
data.head()
```

```python
# Check types
data.dtypes
print(data.isnull().sum())
```

On peut voir que l'on a dans notre dataframe que des variables quantitatives et que notre cible est un Integer dont les valeurs sont compris dans $[0, 1]$ et peut être considérer comme des booleens. Nous n'aurons donc pas besoin de l'encoder.

On peut aussi voir qu'il existe des lignes qui contiennent des valeurs nulles (`NaN`), nous allons donc les traiter avant de pouvoir d'appliquer un modèle

## Traitement des données
Nous allons commencer par analyser puis remplir les données qui sont nulles (`NaN`).  
On peut voir que seul les variables: `ph`, `Sulfate` et `Trihalomethanes` contiennent des valeurs nulles

```python
# Case 1: Removing all missing data
if CASE==0:
    data = data.dropna(how = 'any', axis = 'rows')
elif CASE==1:
# Case 2: Replace NaN values by interpolation then drop the remaining rows containing a NaN value
    data = data.interpolate(method ='nearest', limit_direction ='forward')
    data = data.interpolate(method ='linear', limit_direction ='forward')
    data = data.dropna(how = 'any', axis = 'rows')
else:
# Case 3: Replace NaN values by 
    data['ph']=data['ph'].fillna(data.groupby(['Potability'])['ph'].transform('mean'))
    data['Sulfate']=data['Sulfate'].fillna(data.groupby(['Potability'])['Sulfate'].transform('mean'))
    data['Trihalomethanes']=data['Trihalomethanes'].fillna(data.groupby(['Potability'])['Trihalomethanes'].transform('mean'))
print(data.isnull().sum())
```

On peut donc voir qu'il n'y a plus de valeur nulles et qu'elles ont été remplacés ou supprimés. On va donc maintenant passer à l'analyse des données et les traiter

```python
data.describe()
```

```python
# water potability counts
print(data['Potability'].value_counts())
sns.countplot(x="Potability", data=data)
count_data=data.count()
data.shape
```

On peut remarquer que `potability` est deja encode:
- `0`: eau non potable (**1200** valeurs)
- `1`: eau potable (**811** valeurs)

pour un total de **2011** valeurs valides dans un dataset de **3276** valeurs  
Il serait alors pertinent de remplacer les valeurs nulles par une valeur dont le nombre d'occurence est max.

Nous avons donc ensuite remplacer toutes les valeurs nulles par interpolation ou par la moyenne de cette colonne et nous avons donc:
- `0`: **1998** valeurs
- `1`: **1278** valeurs
pour un total d'environ **3276** valeurs, soit presque l'intégralité des données

## Modèle regression logistique
Le but de notre projet est de prédire si une eau est potable ou non et nous voulons donc une classification de l'eau.  
Après avoir réaliser un traitement de nos données, nous avons étudié la matrice de corrélation pour ne selectionner que les variables dont la corrélation est pertinente et ensuite appliquer le dataset sur un model de regression logistique.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
# Solution of the problem
Y = data['Potability']
# Variables
X = data.drop(columns = ['Potability', 'Trihalomethanes', 'Conductivity'], axis = 1)

# Separate the dataset into two subset: train(80%) and test (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle=True, train_size=.8, random_state=42)
```

```python
cor.Potability.abs().sort_values(ascending = False)[1:]
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
logreg.score(X_test, Y_test)
```

```python
logreg_preds = logreg.predict(X_test)
accuracy = accuracy_score(Y_test, logreg_preds)
mean_accuracy = (cross_val_score(estimator=logreg, X=X_train, y=Y_train, cv=5, scoring='accuracy').mean())
print('Coefficients: ', logreg.coef_)
print('Valid Accuracy (%):', accuracy * 100)
print('Accuracy (%): ', mean_accuracy * 100)
print(classification_report(Y_test, logreg_preds))
```

```python
precision_recall_fscore_support(Y_train, logreg.predict(X_train), average="binary")
```

```python
cm = confusion_matrix(Y_test, logreg_preds)
plt.figure(figsize=(9,9))
sns.heatmap(cm/np.sum(cm), annot=True, fmt=  '0.2%', linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {:.2f}'.format(accuracy)
plt.title(all_sample_title, size = 15);
```

On peut remarquer que ce modèle n'est pas performant, en effet on a une accuracy de **~57%** (puis 63% après remplissage des valeurs nulles), mais surtout en observant la matrice de confusion que tous les positifs n'ont pas été prédit correctement

## Arbre de decision
Afin d'avoir une meilleure visualisation, nous allons utiliser un arbre de decision et calibrer les hyperparamètres afin d'avoir un `accuracy` optimal sans tombé dans le cas de *sur-entrainement* ni de *sous-entrainement*, pour cela nous allons tout d'abord limiter la hauteur et le nombre de feuille de notre arbre de decision.

```python
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier(max_leaf_nodes=15 , max_depth=5, random_state=42, splitter='best')
decision_tree.fit(X_train, Y_train)
decision_tree.score(X_test, Y_test)
```

```python
plt.figure(figsize=(19,9))
tree.plot_tree(decision_tree) 
plt.show()
```

```python
y_pred = decision_tree.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```python
cm = confusion_matrix(Y_test, np.array(y_pred))
plt.figure(figsize=(9,9))
sns.heatmap(cm/np.sum(cm), annot=True, fmt=  '0.2%', linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {:.2f}'.format(accuracy_score(Y_test, y_pred))
plt.title(all_sample_title, size = 15);
```

## Random forest
Ayant des variables independantes et que ses valeurs sont eparpillés, nous allons ensuite utiliser le `Random forest`

```python
from sklearn.ensemble import RandomForestClassifier

random = RandomForestClassifier(max_depth=85,
                                n_estimators=600,
                                n_jobs=-1)
random.fit(X_train, Y_train)
random.score(X_test, Y_test)
```

```python
logreg_random = random.predict(X_test)
print(classification_report(Y_test, logreg_random))
```

```python
rank = random.feature_importances_
std = np.std([tree.feature_importances_ for tree in random.estimators_], axis=0) # On mesure la variance des caracs
forest_importances = pd.Series(rank, index=X.columns)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Carac importances")
ax.set_ylabel("Mean accuracy decrease")
fig.tight_layout()
```

On peut voir que le `Sulfate` et le `ph` sont des variables qui sont importantes

```python
cm = confusion_matrix(Y_test, np.array(logreg_random))
plt.figure(figsize=(9,9))
sns.heatmap(cm/np.sum(cm), annot=True, fmt=  '0.2%', linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {:.2f}'.format(accuracy_score(Y_test, logreg_random))
plt.title(all_sample_title, size = 15);
```

```python
from sklearn.metrics import plot_roc_curve, plot_det_curve

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(22,8))
ax=ax.flatten()
# plot ROC curve
plot_roc_curve(random, X_test, Y_test, ax=ax[0]) 
plot_roc_curve(decision_tree, X_test, Y_test, ax=ax[0]) 
plot_roc_curve(logreg, X_test, Y_test, ax=ax[0]) 
ax[0].set_title("Receiver Operating Characteristic (ROC) curves")
ax[0].plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
        label='Chance', alpha=.8)
ax[0].legend(loc="lower right")

# plot DET curve
plot_det_curve(random, X_test, Y_test, ax=ax[1])
plot_det_curve(decision_tree, X_test, Y_test, ax=ax[1]) 
plot_det_curve(logreg, X_test, Y_test, ax=ax[1])
ax[1].set_title("Detection error tradeoff (DET) curves")
plt.show()
```

On peut voir dans les graphes ci dessus que l'algorithme de Random Forest est bien le plus précis par l'allure des courbes et que la courbe de `DecisionTreeClassifier` et de `RandomForestClassifier` ont presque la même allure dans les deux courbes

## Conclusion
En comparant différent modèle et en traitant les données de façon différente, on est passé d'une `accuracy` de ~ 60% à une accuracy de **~80%** avec un taux de vrai positif de ~26%. Ce qui n'est pas mal étant donné le jeu de donnée que l'on a mais peut être insuffisant étant donné la problèmatique qui est un enjeu majeur pour l'homme.  
On pourrait peut être améliorer l'`accuracy` en traitant les valeurs nulles et les valeurs aberrantes (qui ont été supprimé au cours de ce projet) plus efficacement, par exemple en comparant la ligne contenant la valeur nulle avec les autres lignes et l'affecter à la valeur de la ligne qui possède le maximum d'occurence de similitude.
